```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

spacex_df = pd.read_csv("dataset_part_2.csv")
launch_sites_df = spacex_df[["LaunchSite", "Latitude", "Longitude"]].drop_duplicates()

site_map = folium.Map(location=[29.559684888503615, -95.0830971930759], zoom_start=4)

marker_cluster = MarkerCluster()

for _, site in launch_sites_df.iterrows():
    marker = folium.Marker(
        location=[site["Latitude"], site["Longitude"]],
        popup=site["LaunchSite"],
        icon=folium.Icon(color="red", icon="info-sign")
    )
    marker_cluster.add_child(marker)

site_map.add_child(marker_cluster)
site_map
```

```python
site_map.save("spacex_launch_site_map.html")
```
