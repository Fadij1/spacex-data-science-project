```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("dataset_part_2.csv")
df.head()
```

```python
sns.catplot(x="LaunchSite", y="Class", data=df, kind="bar", aspect=2)
plt.title("Success Rate by Launch Site")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect=2)
plt.title("Flight Number vs Launch Site")
plt.show()
```

```python
sns.catplot(y="LaunchSite", x="PayloadMass", hue="Class", data=df, aspect=2)
plt.title("Payload Mass vs Launch Site")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Orbit", hue="Class")
plt.title("Orbit vs Success")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(x="FlightNumber", y="Orbit", hue="Class", data=df, aspect=2)
plt.title("Flight Number vs Orbit")
plt.show()
```

```python
sns.catplot(x="PayloadMass", y="Orbit", hue="Class", data=df, aspect=2)
plt.title("Payload Mass vs Orbit")
plt.show()
```

```python
success_by_year = df.groupby("Year")["Class"].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=success_by_year, x="Year", y="Class")
plt.title("Success Rate Over Years")
plt.ylabel("Success Rate")
plt.show()
```
