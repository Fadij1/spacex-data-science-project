```python
import pandas as pd
import numpy as np

df = pd.read_csv("dataset_part_1.csv")

df.head()
df.isnull().sum()
```

```python
df["PayloadMass"] = df["PayloadMass"].fillna(df["PayloadMass"].mean())
df["Block"] = df["Block"].fillna(df["Block"].mode()[0] if not df["Block"].mode().empty else 0)
df["Flights"] = df["Flights"].fillna(1)
df["GridFins"] = df["GridFins"].fillna(False)
df["Reused"] = df["Reused"].fillna(False)
df["Legs"] = df["Legs"].fillna(False)

df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
```

```python
features = df[[
    "FlightNumber",
    "PayloadMass",
    "Orbit",
    "LaunchSite",
    "Flights",
    "GridFins",
    "Reused",
    "Legs",
    "Block",
    "Longitude",
    "Latitude",
    "Class",
    "Year"
]]

features.head()
```

```python
features.to_csv("dataset_part_2.csv", index=False)
```
